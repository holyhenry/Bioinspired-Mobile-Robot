```python
import numpy as np
import matplotlib.pyplot as plt

w = 0.6*10**(-3) #m
h = 0.5*10**(-3) #m
L = 50*10**(-3)  #m
rho = 29.4*10**(-8)

ep = np.arange(0, 101)/100
# print(ep)

r = np.zeros(len(ep))

for i in range(len(ep)):
    r[i] = rho*ep[i]*L*(8-ep[i])/(w*h*(2-ep[i])**2)
#     print(r[i])
    
plt.plot(ep,r)
plt.xlabel('strain(%)')
plt.ylabel('Risistance change')
plt.grid()
plt.show()
```
